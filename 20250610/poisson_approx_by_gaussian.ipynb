```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as ss
```

```python
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 75
```

```python
# show that Poisson distribution is approximated really well by gaussian
# for mu large
mu = 150; imax = int(2 * mu); sigma = np.sqrt(mu)
xs = np.linspace(0, 2 * mu, 1000)
plt.plot(range(imax), [ss.poisson(mu).pmf(i) for i in range(imax)], "+", label="Poisson (exact)")
plt.plot(xs, 1.0/np.sqrt(2*np.pi)/sigma * np.exp(-0.5 * ((xs-mu)/sigma)**2),label="gaussian (approx)")
plt.gca().legend()
```
