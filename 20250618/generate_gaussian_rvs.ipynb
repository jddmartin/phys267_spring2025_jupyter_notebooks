```python
# example of generating random variables according to Gaussian distribution
# with mu=0, sigma=1
```

```python
import random
import matplotlib.pyplot as plt
import scipy.stats as ss
import scipy.optimize as so
```

```python
def rv():
    while True:
        y = random.random()
        def f(x): return ss.norm(0,1).cdf(x) - y
        try:
            x = so.brentq(f, -50, 50)  # could fail if x is very close to zero or one
            return x
        except ValueError:
            pass  # try again
```

```python
rvs = [rv() for _ in range(15000)]
```

```python
plt.hist(rvs,bins=20)
```
