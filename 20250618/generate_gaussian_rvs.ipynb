```python
# example of generating random variables according to Gaussian distribution
# with mu=0, sigma=1
```

```python
import random
import matplotlib.pyplot as plt
import scipy.stats as ss
import scipy.optimize as so
import numpy as np
```

```python
def rv():
    while True:
        y = random.random()
        def f(x): return ss.norm(0,1).cdf(x) - y
        try:
            x = so.brentq(f, -50, 50)  # could fail if x is very close to zero or one
            return x
        except ValueError:
            pass  # try again
```

```python
rvs = [rv() for _ in range(15000)]
```

```python
bins = np.linspace(-5.5,5.5,22)
ax = plt.gca()
ax.hist(rvs, bins=bins, density=True, label="generated")
xs = np.linspace(-4, 4, 1000)
ys = 1.0 / np.sqrt(2 * np.pi) * np.exp(-0.5 * xs**2)
ax.plot(xs, ys, label="theory")
ax.legend()
```
