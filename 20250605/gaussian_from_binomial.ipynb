```python
n = 100
p = 0.5
f = ss.binom(n, p).pmf
sigma2 = n * p * (1-p)
```

```python
iss = list(range(0, n+1))
xs = np.linspace(0,100,1000)
def fg(x): return 1/np.sqrt(2*np.pi*sigma2) * np.exp(-0.5 / sigma2 * (x-n*p)**2)
```

```python
plt.plot(iss, f(iss), "+", label="binomial exact")
plt.plot(xs, fg(xs), label="Gaussian approximation")
plt.legend()
```

```python
plt.plot(iss, np.log(f(iss)), "+", label="binomial exact")
plt.plot(xs, np.log(fg(xs)), label="Gaussian approximation")
plt.legend()
```
