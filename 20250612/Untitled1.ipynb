```python
import numpy as np
import scipy.stats as ss
import scipy.optimize as so
import scipy.special as ssf
```

```python
z=1.96; ss.norm(0,1).cdf(z)-ss.norm(0,1).cdf(-z)
```

```python
def f(z): return ss.norm(0,1).cdf(z)-ss.norm(0,1).cdf(-z) - 0.95
```

```python
f(1.96)
```

```python
zr = so.brentq(f, 1.9, 2)
```
