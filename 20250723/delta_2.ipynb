```python
import random
import numpy as np
import matplotlib.pyplot as plt
random.seed(2)
```

```python
def variate_for_exponential(R):
    cdf = random.random()
    return np.log(1 - cdf)/(-R)
```

```python
n = int(1e6); ts = np.array([variate_for_exponential(1) for _ in range(n)])
```

```python
dt2 = ts[0:-1] + ts[1:]
```

```python
plt.hist(dt2, density=True, bins=100)
ax = plt.gca()
ax.set_xlim([0,5])
def f_theory(t): return t * np.exp(-t)
tt = np.linspace(0, 10, 100)
ax.plot(tt, f_theory(tt), color="orange")
```
